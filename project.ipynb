```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder # Import LabelEncoder
from sklearn.metrics import (confusion_matrix, classification_report,
                             accuracy_score, f1_score, roc_auc_score,
                             roc_curve, mean_squared_error)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Load dataset
df = pd.read_csv('healthcare_dataset1 .csv')  # Replace with your filename

# Display basic info
print("Dataset preview:")
print(df.head())
print("\nMissing values:\n", df.isnull().sum())

# Preprocessing
df.dropna(inplace=True)  # Drop missing rows (or use imputation)

# --- FIX ---
# Identify the actual target column name by inspecting df.columns
# For example, if your target is 'Medical Condition', replace 'target' below with 'Medical Condition'
print("DataFrame columns:", df.columns) # Print columns to find the target column name
# Replace 'YOUR_TARGET_COLUMN_NAME' with the actual target column name from the print output
# Assuming 'Medical Condition' is the target variable
target_column_name = 'Medical Condition' # <--- Set the correct target column name

# Separate features (X) and target (y)
X = df.drop(columns=[target_column_name])
y = df[target_column_name]
# --- END FIX ---

# Encode the target variable 'y' if it's categorical (which 'Medical Condition' likely is)
# Check the type of y before encoding
if y.dtype == 'object' or y.dtype == 'category':
    print("\nEncoding target variable...")
    le_y = LabelEncoder()
    y = le_y.fit_transform(y)
    print("Target variable encoded.")
    print("Encoded target classes:", le_y.classes_) # Show the mapping of original to encoded values
else:
    print("\nTarget variable is already numerical.")


# Encode categorical variables in X if any
X = pd.get_dummies(X, drop_first=True)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000), # Increase max_iter for convergence
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(probability=True, random_state=42)
}

# Evaluation results
results = []

for name, model in models.items():
    print(f"\nTraining {name}...")
    # Handle potential multi-class scenario for metrics like f1 and roc_auc
    # If target is multi-class, f1_score needs 'average' parameter
    # roc_auc_score requires either binary classification or multi-class support from the model

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Check if the target is binary for certain metrics
    is_binary = (len(np.unique(y_test)) == 2)

    acc = accuracy_score(y_test, y_pred)

    # F1 score calculation
    if is_binary:
        f1 = f1_score(y_test, y_pred)
    else:
        # Use 'weighted' average for multi-class F1 score
        f1 = f1_score(y_test, y_pred, average='weighted')
        print(f"Warning: F1 score calculated with 'weighted' average for multi-class problem.")


    # ROC AUC calculation
    roc = np.nan # Initialize ROC AUC to NaN
    if is_binary and hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
        roc = roc_auc_score(y_test, y_proba)
    elif not is_binary and hasattr(model, "predict_proba"):
         # For multi-class ROC AUC, you might need One-vs-Rest or similar
         # This is more complex and depends on the problem.
         # For simplicity, we'll skip multi-class ROC AUC here unless specifically needed.
         print(f"Note: Skipping ROC AUC for multi-class classification for {name}.")
         y_proba = None # Ensure y_proba is None if not binary ROC AUC

    else:
        y_proba = None
        print(f"Note: {name} does not have predict_proba or is not a binary classification for ROC AUC.")


    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    results.append({
        'Model': name,
        'Accuracy': acc,
        'F1 Score': f1,
        'ROC AUC': roc,
        'RMSE': rmse
    })

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6)) # Adjust figure size
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le_y.classes_ if 'le_y' in locals() else 'auto', yticklabels=le_y.classes_ if 'le_y' in locals() else 'auto')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve (only for binary classification with predict_proba)
    if is_binary and y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc:.2f})')


# Plot ROC Curves (if any binary models had predict_proba)
if any(r['ROC AUC'] is not np.nan for r in results): # Check if any model had a valid ROC AUC
    plt.figure(figsize=(8, 6))
    for r in results:
        if r['ROC AUC'] is not np.nan:
             # Need to re-compute or store fpr, tpr for plotting all on one graph
             # For simplicity here, we plot within the loop, but a better way is to collect curves
             pass # ROC curves are plotted inside the loop for now

    # Re-plot ROC curves outside the loop for a combined graph
    plt.figure(figsize=(8, 6))
    for name, model in models.items():
        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test)
            # Handle binary vs multi-class ROC plot
            if len(np.unique(y_test)) == 2:
                 y_proba_binary = y_proba[:, 1]
                 fpr, tpr, _ = roc_curve(y_test, y_proba_binary)
                 roc_auc = roc_auc_score(y_test, y_proba_binary)
                 plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
            # Multi-class ROC curve plotting is more involved (e.g., One-vs-Rest)
            # Add multi-class ROC plotting if needed, using metrics like roc_auc_score(..., multi_class='ovr')


    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (Binary Classification)')
    plt.legend()
    plt.show()
else:
    print("\nNo binary classification models with predict_proba found to plot ROC curves.")


# Comparison Table
results_df = pd.DataFrame(results)
print("\nModel Comparison:\n", results_df)
```

```python
# ---------------------- LIBRARIES ----------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder # Import LabelEncoder
from sklearn.metrics import (accuracy_score, f1_score, roc_auc_score,
                             mean_squared_error, confusion_matrix, roc_curve)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Optional: Use XGBoost if available
try:
    from xgboost import XGBClassifier
    xgb_available = True
except ImportError:
    xgb_available = False
    print("XGBoost is not installed. Skipping it.")

# ---------------------- LOAD DATA ----------------------
df = pd.read_csv("healthcare_dataset1 .csv")  # Replace with actual filename
print("Dataset shape:", df.shape)
print("DataFrame columns:", df.columns) # Print columns to verify names

# ---------------------- PREPROCESSING ----------------------
df.dropna(inplace=True)

# --- FIX ---
# Use the actual target column name from the DataFrame.
# Based on previous cells and the traceback, the target column is 'Medical Condition'.
target_column_name = 'Medical Condition' # Correct target column name

X = df.drop(columns=[target_column_name])  # Features - drop the actual target column
y = df[target_column_name]                 # Target - select the actual target column
# --- END FIX ---

# Encode the target variable 'y' if it's categorical
if y.dtype == 'object' or y.dtype == 'category':
    print("\nEncoding target variable...")
    le_y = LabelEncoder() # Initialize a new LabelEncoder for the target variable
    y = le_y.fit_transform(y)
    print("Target variable encoded.")
    print("Encoded target classes:", le_y.classes_) # Show the mapping
else:
    print("\nTarget variable is already numerical.")


# Handle categorical variables in X
X = pd.get_dummies(X, drop_first=True)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ---------------------- MODEL BUILDING ----------------------
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Machine": SVC(probability=True),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

if xgb_available:
    # For multi-class classification, eval_metric='mlogloss' is often appropriate
    models["XGBoost"] = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# ---------------------- MODEL EVALUATION ----------------------
results = []

for name, model in models.items():
    print(f"\nTraining: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Check if target is binary or multi-class
    is_binary = (len(np.unique(y_test)) == 2)

    acc = accuracy_score(y_test, y_pred)

    # F1 score (use 'weighted' for multi-class)
    f1 = f1_score(y_test, y_pred, average='weighted') # Always use weighted for consistency with multi-class

    # ROC AUC (handle binary vs multi-class)
    roc = np.nan
    y_proba = None
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
        if is_binary:
             # For binary, take the probability of the positive class (index 1)
             roc = roc_auc_score(y_test, y_proba[:, 1])
        else:
             # For multi-class, use One-vs-Rest (ovr) strategy
             # Ensure you have the correct le_y object in scope or accessible
             try:
                 roc = roc_auc_score(y_test, y_proba, multi_class='ovr', average='weighted')
                 print(f"Note: ROC AUC calculated with 'ovr' and 'weighted' average for multi-class problem for {name}.")
             except ValueError as e:
                 print(f"Warning: Could not calculate multi-class ROC AUC for {name}. Error: {e}")
                 roc = np.nan # Set to NaN if calculation fails (e.g., single class in test set)
    else:
        print(f"Note: {name} does not have predict_proba. Skipping ROC AUC.")


    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    results.append({
        'Model': name,
        'Accuracy': acc,
        'F1 Score': f1,
        'ROC AUC': roc,
        'RMSE': rmse
    })

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(len(np.unique(y_test)) * 1.5, len(np.unique(y_test)) * 1.5)) # Dynamic size
    # Use le_y.classes_ for tick labels if encoding was done
    xticklabels = le_y.classes_ if 'le_y' in locals() else 'auto'
    yticklabels = le_y.classes_ if 'le_y' in locals() else 'auto'
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=xticklabels, yticklabels=yticklabels)
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve (Plotting within the loop for individual plots)
    # Combined ROC plot is handled outside the loop
    # We can plot binary ROC here, or collect data for the combined plot

# Final ROC plot (Combined)
plt.figure(figsize=(8, 6))
has_roc_plot = False # Flag to check if any ROC curve was plotted

for name, model in models.items():
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
        is_binary = (len(np.unique(y_test)) == 2)

        if is_binary:
            # For binary, take the probability of the positive class
            y_proba_binary = y_proba[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_proba_binary)
            roc_auc = roc_auc_score(y_test, y_proba_binary)
            plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
            has_roc_plot = True
        else:
             # Multi-class ROC curves are more complex and not directly a single line.
             # Skipping combined multi-class ROC plot for simplicity.
             pass


if has_roc_plot:
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Comparison (Binary Classification Only)') # Indicate if only binary
    plt.legend()
    plt.show()
else:
    print("\nNo binary classification models with predict_proba found to plot combined ROC curves.")


# ---------------------- COMPARISON TABLE ----------------------
results_df = pd.DataFrame(results)
print("\n📊 Model Comparison:\n", results_df.sort_values(by="F1 Score", ascending=False))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import gradio as gr

# Load dataset
df = pd.read_csv("healthcare_dataset1 .csv")

# Select and clean data
df_model = df[['Age', 'Gender', 'Blood Type', 'Medical Condition']].dropna()

# Encode target
label_encoder = LabelEncoder()
df_model['Medical Condition'] = label_encoder.fit_transform(df_model['Medical Condition'])

# Define features and target
X = df_model[['Age', 'Gender', 'Blood Type']]
y = df_model['Medical Condition']

# Preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age']),
        ('cat', OneHotEncoder(), ['Gender', 'Blood Type'])
    ])

# Model pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
pipeline.fit(X_train, y_train)

# Prediction function
def predict_condition(age, gender, blood_type):
    input_data = pd.DataFrame([[age, gender, blood_type]], columns=['Age', 'Gender', 'Blood Type'])
    pred = pipeline.predict(input_data)[0]
    return label_encoder.inverse_transform([pred])[0]

# Gradio interface
gr.Interface(
    fn=predict_condition,
    inputs=[
        gr.Number(label="Age"),
        gr.Radio(["Male", "Female"], label="Gender"),
        gr.Radio(["A+", "A-", "B+", "B-", "AB+", "AB-", "O+", "O-"], label="Blood Type")
    ],
    outputs="text",
    title="Disease Predictor",
    description="Enter Age, Gender, and Blood Type to predict the medical condition."
).launch()
```
